```python
##################################################################################
##################################################################################
#                   ŸÖÿ≠ŸÖÿØ ÿßŸäŸÖŸÜ ŸÅÿ™ÿ≠Ÿä ÿπÿ®ÿØ ÿßŸÑÿ∫ŸÜŸä ÿßŸÑÿ≥ŸäÿØ                            #
##################################################################################
##################################################################################
car_store = {
    "Toyota Corolla": 1250000,
    "Nissan Sunny": 670000,
    "Hyundai Elantra": 920000,
    "Kia Sportage": 1550000,
    "MG ZS": 900000,
    "Chevrolet Optra": 700000
}

cart = []

def show_menu():
    print("\n Welcome to EL-Halmoushy_Auto \n")
    print("1. View available cars")
    print("2. Add a car to your cart(want to buy)")
    print("3. View cart and total cost")
    print("4. Exit\n\n\n\n")

while True:
    show_menu()
    choice = input("Choose an option (1‚Äì4): ")

    if choice == "1":
        print("\nAvailable Cars for Sale:")
        for car, price in car_store.items():
            print(f"- {car}: {price:,} EGP")

    elif choice == "2":
        car = input("Enter the full car name (as: Kia Sportage ): ")
        if car in car_store:
            cart.append(car)
            print(f"\n{car} has been added to your cart.\n")
        else:
            print("Sorry, this car is not available.")

    elif choice == "3":
        if not cart:
            print(" Your cart is empty.")
        else:
            print("\n Your Selected Cars:")
            total = 0
            for car in cart:
                price = car_store[car]
                print(f"- {car}: {price:,} EGP")
                total += price
            print(f"\n Total Price: {total:,} EGP")

    elif choice == "4":
        print("Thank you for visiting EL-Halmoushy_Auto !")
        break

    else:
        print("Invalid input. Please select from the menu.")
##################################################################################
##################################################################################
#                   ŸÖÿ≠ŸÖÿØ ÿßŸäŸÖŸÜ ŸÅÿ™ÿ≠Ÿä ÿπÿ®ÿØ ÿßŸÑÿ∫ŸÜŸä ÿßŸÑÿ≥ŸäÿØ                            #
##################################################################################
##################################################################################
```
